```python
#lista desde el 1 al 100
for x in range(1, 101):
    if x % 10 == 0:
        print(x, "Bip Bip")
    elif x % 5 == 0:
        print(x, "Bip")
```

```python
# Solicitar dos números al usuario
numero1 = float(input("Ingresa el primer número: "))
numero2 = float(input("Ingresa el segundo número: "))

# Sumar los números
suma = numero1 + numero2

# Mostrar el resultado de la suma
print("La suma es:", suma)

# Evaluar el nivel
if suma < 50:
    print("Nivel principiante")
elif suma < 80:
    print("Nivel intermedio")
else:
    print("Nivel pro")
```

```python
# Solicitar datos
edad = int(input("Ingresa tu edad: "))
le_gusta_musica = input("¿Te gusta la música? (si/no): ").lower()
le_gusta_bailar = input("¿Te gusta bailar? (si/no): ").lower()

# Convertir respuestas a booleanos
gusta_musica = le_gusta_musica == "si"
gusta_bailar = le_gusta_bailar == "si"

# Evaluar perfil
if edad >= 18 and gusta_musica and gusta_bailar:
    print("Eres una estrella de la fiesta")
elif gusta_musica and not gusta_bailar:
    print("Te gusta la música, pero no tanto bailar")
else:
    print("Tienes tu propio estilo")
```
